```python
#Import the Pandas library
import pandas as pd
#Read the CSV data in the path provided
data = pd.read_csv(r"C:\Users\Dennis\Desktop\VISUALIZATION DATA\New folder\employee_data.csv")
#Print the data to view it
data
```

```python
# Convert the 'StartDate' and 'ExitDate' columns to datetime
data['StartDate'] = pd.to_datetime(data['StartDate'])
data['ExitDate'] = pd.to_datetime(data['ExitDate'])

# Calculate the tenure for each employee by subtracting 'StartDate' from 'ExitDate',
# and then dividing the result by the number of days in a year (365).
data['Tenure'] = (data['ExitDate'] - data['StartDate']).dt.days/ 365

#Calculate the average tenure of all employees
Average_tenure = data['Tenure'].mean()

#Print the average tenure in years of employees
print('The average tenure in years of employees is:',Average_tenure)
```

```python
#Import the visualization library
import plotly.express as px

#Calculate the counts of each unique value in the 'EmployeeStatus' column.
Status_counts = data['EmployeeStatus'].value_counts()

# Create a pie chart using Plotly Express.
#'names' parametre specifies the labels for the segment.
#'values' parametre specifies the values for each segment.
fig = px.pie(Status_counts, names=Status_counts.index ,values=Status_counts.values)

#Updatethe pie chart's traces to show both the percentage and label on the chart.
fig.update_traces(textinfo='percent+label')

#Display the pie chart
fig.show()

fig.write_image("Employee status.png")
```

There are 2458 employees who are currently working (active)

#**DURATION OF EMPLOYMENT IN DAYS**

```python
import pandas as pd
import datetime
# Convert the 'DOB' column to datetime format.
data['DOB'] = pd.to_datetime(data['DOB'])

# Calculate the age for each employee by subtracting their birth year from the current year.
data['Age'] = (datetime.datetime.now().year - data['DOB'].dt.year)

# Group the data by 'DepartmentType' and calculate the average age for each group.
Average_age_by_department = data.groupby('DepartmentType')['Age'].mean()
Average_age_by_department
```
